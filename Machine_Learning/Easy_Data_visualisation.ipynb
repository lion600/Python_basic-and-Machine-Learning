```python
import matplotlib.pyplot as plt
import numpy as np

# Generating dummy data
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)

# Creating a scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Dummy Data')
plt.title('Dummy Data Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 35]

plt.figure(figsize=(8, 6))
plt.bar(categories, values, color='skyblue')
plt.title('Bar Chart Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart Example')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generating random data for the box plot
np.random.seed(42)
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.figure(figsize=(8, 6))
plt.boxplot(data, patch_artist=True)
plt.title('Box Plot Example')
plt.xlabel('Data')
plt.ylabel('Value')
plt.xticks(ticks=[1, 2, 3], labels=['A', 'B', 'C'])
plt.grid(axis='y')
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt

# Load Breast Cancer dataset
cancer = load_breast_cancer()
data = cancer.data[:, 0]

cancer
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of First Feature (Breast Cancer Dataset)')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
diagnosis_count = [sum(cancer.target == 0), sum(cancer.target == 1)]
diagnosis_labels = cancer.target_names

plt.figure(figsize=(8, 6))
plt.bar(diagnosis_labels, diagnosis_count, color=['skyblue', 'salmon'])
plt.title('Diagnosis Count (Breast Cancer Dataset)')
plt.xlabel('Diagnosis')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
```

```python
diagnosis_count = [sum(cancer.target == 0), sum(cancer.target == 1)]
diagnosis_labels = cancer.target_names

# Creating a pie chart
plt.subplot(1, 2, 2)
plt.pie(diagnosis_count, labels=diagnosis_labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'salmon'])
plt.title('Diagnosis Count (Pie Chart)')

plt.tight_layout()
plt.show()
```

```python
data = [cancer.data[cancer.target == 0][:, 0], cancer.data[cancer.target == 1][:, 0]]

plt.figure(figsize=(8, 6))
plt.boxplot(data, patch_artist=True, notch=True, showmeans=True)
plt.title('Boxplot of First Feature (Breast Cancer Dataset)')
plt.xlabel('Diagnosis (0: Malignant, 1: Benign)')
plt.ylabel(f'{cancer.feature_names[0]}')
plt.xticks([1, 2], cancer.target_names)
plt.grid(axis='y')
plt.show()
```

```python
import seaborn as sns
selected_features = [0, 3, 6, 20, 27]
selected_feature_names = [cancer.feature_names[i] for i in selected_features]
selected_data = cancer.data[:, selected_features]

corr_matrix = np.corrcoef(selected_data.T)

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', xticklabels=selected_feature_names, yticklabels=selected_feature_names)
plt.title('Correlation Heatmap of Selected Features (Breast Cancer Dataset)')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```

```python
feature_x = 0
feature_y = 3
x = cancer.data[:, feature_x]
y = cancer.data[:, feature_y]

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=cancer.target, cmap='viridis', edgecolor='black')
plt.colorbar(label='Diagnosis', ticks=[0, 1], format=plt.FuncFormatter(lambda val, _: cancer.target_names[val]))
plt.title('Scatter Plot of Two Features (Breast Cancer Dataset)')
plt.xlabel(f'Feature {feature_x + 1}: {cancer.feature_names[feature_x]}')
plt.ylabel(f'Feature {feature_y + 1}: {cancer.feature_names[feature_y]}')
plt.grid(True)
plt.show()
```
